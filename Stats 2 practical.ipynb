```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab

def plot_data(data):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(data,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(data,dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(lognormal_distributed_data)
# this is a q-q plot right one
```

```python
plot_data(np.log(lognormal_distributed_data))
```

```python
# Central limit theoram -> mean of all the given samples of a population    
```

```python
population=np.random.randint(10,20,50)
```

```python
population
```

```python
np.mean(population)
```

```python
np.median(population)
```

```python
from statistics import mode
mode(population)
```

```python
# random sample form the population

sample=np.random.choice(population,20)
```

```python
print(f"mean of the sample: {np.mean(sample)}")
print(f"median of the sample: {np.median(sample)}")
print(f"mode of the sample: {mode(sample)}")
```

```python
sample1=np.random.choice(population,20)
sample2=np.random.choice(population,20)
sample3=np.random.choice(population,20)
sample4=np.random.choice(population,20)
```

```python
print(sample1)
print(60*'=')
print(sample2)
print(60*'=')
print(sample3)
print(60*'=')
print(sample4)
print(60*'=')
```

```python
mean_of_the_sample=[]

all_sample=[sample1,sample2,sample3,sample4]

for sample in all_sample:
    print(np.mean(sample))
    mean_of_the_sample.append(np.mean(sample))
```
